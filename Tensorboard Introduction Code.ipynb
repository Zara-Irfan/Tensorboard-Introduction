```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")
print(f"Shape of one image: {X_train[0].shape}")
```

```python
plt.matshow(X_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis('off')
plt.show()
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, input_shape=(784,), activation='relu'),  # Hidden layer
    keras.layers.Dense(10, activation='softmax')                      # Output layer for 10 classes
])
```

Compile the Neural Network Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train_flattened, y_train, epochs=5, batch_size=32)
```

```python
test_loss, test_accuracy = model.evaluate(X_test_flattened, y_test)
print(f"Test Accuracy: {test_accuracy*100:.2f}%")
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = np.argmax(y_predicted, axis=1)
```

```python
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.matshow(X_test[i], cmap='gray')   # remove fignum
    plt.title(f"True: {y_test[i]}\nPred: {y_predicted_labels[i]}")
    plt.axis('off')
plt.show()
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("MNIST Confusion Matrix")
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import os

# Dummy data
X_train = np.random.random((100, 28, 28))
y_train = np.random.randint(0, 10, 100)

# Model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Callback
tb_callback = tf.keras.callbacks.TensorBoard(log_dir="logs/run1", histogram_freq=0)

# Compile and train
model.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=2, callbacks=[tb_callback])

# Check event files
print("Event files:", os.listdir("logs/run1"))
```
